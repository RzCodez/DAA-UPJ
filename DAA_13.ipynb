```python
def bfs(graph, start):
    visited = []
    queue = []
    queue = [start]

    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            neighbours = graph[node]
            for neighbour  in neighbours:
                queue.append(neighbour)
    return visited

```

DFS (Depth-First Search) merupakan alternatif dari BFS (Breadth-First Search) yang digunakan untuk mencari data dari sebuah graf.
Perbedaan utama antara DFS dan BFS adalah bahwa setelah memulai dari simpul akar (root vertex), algoritma DFS akan menelusuri simpul sedalam mungkin pada setiap jalur tunggal yang unik secara berurutan.
Untuk setiap jalur, setelah mencapai kedalaman maksimal, algoritma akan menandai semua simpul yang terkait dengan jalur tersebut sebagai telah dikunjungi.


```python
def dfs(visited, graph, node):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbor in graph[node]:
            dfs(visited, graph, neighbor)
```

## Perbandingan BFS dan DFS dari ketiga data graph
**Analisis:**

##BFS

1. **Graf Data Amin:**
   - **Output BFS:** `['Amin', 'Wasim', 'Nick', 'Mike', 'Imran', 'Faras']`
   - **Analisis:** BFS melakukan penelusuran per tingkat, menjelajahi semua tetangga suatu simpul sebelum beralih ke tingkat berikutnya. Dalam hal ini, dimulai dari 'Amin' dan mengunjungi tetangganya ('Wasim', 'Nick', 'Mike') sebelum beralih ke tingkat berikutnya ('Imran', 'Faras'). Hasilnya adalah penelusuran graf secara berdasarkan tingkat.

2. **Graf Data Rektor:**
   - **Output BFS:** `['Rektor', 'Warek 1', 'Warek 2', 'Kaprodi 1', 'Kaprodi 2', 'Kaprodi 3', 'Dosen A', 'Dosen B', 'Dosen C', 'Dosen D', 'Dosen E', 'Dosen F', 'Dosen G']`
   - **Analisis:** Serupa dengan graf Amin, BFS menjelajahi graf per tingkat. Dimulai dari 'Rektor' dan beralih ke tetangganya ('Warek 1', 'Warek 2') sebelum menjelajahi tingkat berikutnya ('Kaprodi 1', 'Kaprodi 2', 'Kaprodi 3', dan dosen terkait).

3. **Graf Data Titik 0:**
   - **Output BFS:** `['0', '9', '7', '11', '10', '8', '3', '6', '1', '12', '2', '4', '5']`
   - **Analisis:** Penelusuran BFS pada graf 'Titik 0' juga mengikuti pola penjelajahan berdasarkan tingkat. Dimulai dari '0' dan mengunjungi tetangganya ('9', '7', '11') sebelum beralih ke tingkat berikutnya, menghasilkan urutan yang diberikan.

##DFS

1. **Graf Data Amin:**
   - **Output DFS:** `Amin, Wasim, Imran, Faras, Nick, Mike`
   - **Analisis:** DFS menjelajahi sejauh mungkin sepanjang setiap cabang sebelum melakukan backtracking. Dalam hal ini, dimulai dari 'Amin', menjelajahi sejauh mungkin sepanjang setiap cabang sebelum melakukan backtracking. Hasilnya adalah penelusuran graf secara berdasarkan kedalaman.

2. **Graf Data Rektor:**
   - **Output DFS:** `Rektor, Warek 1, Warek 2, Kaprodi 1, Dosen A, Dosen B, Dosen C, Kaprodi 2, Dosen D, Dosen E, Kaprodi 3, Dosen F, Dosen G`
   - **Analisis:** DFS dimulai dari 'Rektor' dan menjelajahi setiap cabang sejauh mungkin sebelum melakukan backtracking. Hasilnya adalah penelusuran graf secara berdasarkan kedalaman.

3. **Graf Data Titik 0:**
   - **Output DFS:** `0, 1, 8, 9, 7, 10, 11, 6, 5, 3, 2, 4`
   - **Analisis:** DFS dimulai dari '0' dan menjelajahi sejauh mungkin sepanjang setiap cabang sebelum melakukan backtracking. Hasilnya adalah penelusuran graf secara berdasarkan kedalaman.

**Kesimpulan:**

BFS Cocok untuk menemukan lintasan terpendek dalam graf tanpa bobot, penelusuran per tingkat, sedangkan DFS cocok untuk pengurutan topologis dan menemukan komponen terhubung, menjelajahi sejauh mungkin sebelum melakukan backtracking.

Pilihan antara BFS dan DFS tergantung pada kasus data graphnya. Jika tujuannya adalah menemukan path terpendek atau mengeksplor tetangga per tingkat, BFS lebih cocok. Jika tujuannya adalah mengeksplor sejauh mungkin, DFS adalah pilihan yang lebih baik.
